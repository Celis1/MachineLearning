## fixing columns

```python
df = pd.read_csv('./train.csv')
```

#### creating family column

```python
#creating new column of family summed up
df['FamilyCount'] = df[['SibSp','Parch']].sum(axis = 1)
```

```python
df.head()
```

#### fixing age

```python
df.Age.isna().sum()
```

```python
df.Age.fillna(value=df.Age.mean(),inplace=True)
```

```python
df['Age'] = [int(x) for x in df.Age]
```

```python
df.head()
```

#### looking into cabins

```python
df.Cabin.dropna()
```

```python
df.Cabin
```

```python
#looking at all the uniques first letters 
first_val = [x[:1] for x in df.Cabin.dropna() ]
pd.Series(first_val).unique()
```

```python
cabins = []

#filling all the nulls with a U for unkown
df.Cabin.fillna(value=('U'),inplace=True)

#adds the first letter 
for i in df.Cabin:
    cabins.append(str(i)[:1])
        
df['Cabin'] = cabins


cabin_list = {'C': 0, 'E': 1, 'G': 2, 'D':3, 'A':4, 'B':5, 'F':6, 'T':7,'U':8}

df['Cabin'] = df.Cabin.map(cabin_list)
```

```python
df.head()
```

#### fixing fare

```python
df.Fare
```

```python
df['Fare'] = [int(x) for x in df.Fare]
```

```python
df.head()
```

#### fixing Embarked

```python
df.Embarked
```

```python
df.Embarked.fillna('S')
ports = {'S':0 ,'C':1 ,'Q':2 }

df['Embarked'] = df.Embarked.map(ports)
```

```python
df.head()
```

#### fixing names

```python
df.Name
```

```python
names = []

for i in df.Name:
    names.append(i[i.index(',')+2:i.index('.')])
df['Name'] = names
```

```python
df.Name.unique()
```

```python
name_list = {'Mr':0, 'Mrs':2, 'Miss':1, 'Master':3, 'Don':4, 'Rev':5, 'Dr':6, 'Mme':1, 'Ms':1,
       'Major':7, 'Lady':8, 'Sir':9, 'Mlle':1, 'Col':10, 'Capt':13, 'the Countess':11,
       'Jonkheer':12}

df['Name'] = df.Name.map(name_list)
```

```python
df.head()
```

# cleaning data

```python
df.head()
```

```python
#creating test df 
test_df = df.copy()
```

```python
df.columns
```

```python
#dropping unwanted columns
def drop_cols(df):
    #picking columns i want
    cols = [
     'PassengerId',
      'Survived', 
      'Pclass',
      'Name', 
      'Male', 
      'Age', 
      'SibSp',
      'Parch',
#      'Ticket',
      'Fare', 
      'Cabin',
      'Embarked',
      'FamilyCount'
    ]
    
    
    df = df[cols]
    return(df)
```

```python
test_df = drop_cols(df)
```

```python
test_df.head()
```

```python
test_df.isnull().sum()
```

```python
test_df = test_df.dropna()
test_df.head(7)
```

```python
sns.heatmap(df.corr(),annot=True)
```

## using model to predict

```python
from sklearn.model_selection import train_test_split,cross_val_score
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LinearRegression,LogisticRegressionCV
```

### splitting the targeted columns

```python
#creating the target variable 
y = test_df['Survived']
X = test_df.drop(columns=['Survived'])
```

```python
print(y.shape)
X.head()
```

```python
#creating train test split
x_train,x_test,y_train,y_test = train_test_split(X,y)
```

#### Logistic Regression

```python
lr = LogisticRegressionCV()
lr.fit(x_train,y_train)
print('training score:' ,cross_val_score(lr,x_train,y_train).mean())
print('test score: ', cross_val_score(lr,x_test,y_test).mean())
```

```python
lr.score(x_train,y_train)
```

#### Random Forest

```python
from sklearn.grid_search import GridSearchCV
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier(n_estimators=40,min_samples_split=5)
rf.fit(x_train,y_train)
cross_val_score(rf,x_train,y_train).mean()
```
